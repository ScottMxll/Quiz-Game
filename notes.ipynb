It seems that when it comes to ':=' operator it can only be used inside of a function, it cannot be used to redeclare an existing variable in the same scope, but alas it can be used to initialize multiple variables at once, 
like this:

> name, age := "Siinzy", 27

The %v is a format specifier for values & is known as the 'format verb'. Its used inside functions like 'fmt.Printf' & 'fmt.Sprintf'. It works with, strings, integers, floats, booleans, structs, arrays, and slices.
